```python
#Tarvittavien kirjastojen tuominen.
import pandas as pd
```

```python
#Tallensin datan koneelleni ja avasin sen.
df=pd.read_excel("kunnat-1.xlsx", sheet_name="Kuntien avainluvut")
```

```python
#Tässä kohdassa olen lyhentänyt muutamien muuttujien nimiä.
df.columns=['Kunta', 'Taajama-aste, %, 2017', 'Väkiluku 2017',
       'Väkiluvun %-muutos edellisestä vuodesta 2017',
       'Alle 15v %-osuus väestöstä 2017',
       '15-64v %-osuus väestöstä 2017',
       'Yli 64v %-osuus väestöstä 2017',
       'Ruotsinkielisten %-osuus väestöstä 2017',
       'Ulkomaan kansalaisten %-osuus väestöstä 2017',
       'Syntyneiden enemmyys, henkilöä, 2017',
       'Kuntien välinen muuttovoitto/-tappio, henkilöä, 2017',
       'Perheiden lukumäärä 2017', 'Asuntokuntien lukumäärä 2017',
       'Rivi- ja pientaloissa asuvien asuntokuntien %-osuus 2017',
       'Vuokra-asunnoissa asuvien asuntokuntien %-osuus 2017',
       'Vähintään toisen asteen tutkinnon suorittaneiden %-osuus 15v täyttäneistä 2017',
       'Korkea-asteen tutkinnon suorittaneiden %-osuus 15v täyttäneistä 2017',
       'Alueella asuvan työllisen työvoiman määrä, 2016',
       'Työllisyysaste, %, 2016',
       'Asuinkunnassaan työssäkäyvien %-osuus 2016',
       'Työttömien %-osuus työvoimasta 2016',
       'Eläkeläisten %-osuus väestöstä 2016',
       'Taloudellinen huoltosuhde 2016',
       'Alueella olevien työpaikkojen lukumäärä 2016',
       'Alkutuotannon työpaikkojen %-osuus 2016',
       'Jalostuksen työpaikkojen %-osuus 2016',
       'Palvelujen työpaikkojen %-osuus 2016',
       'Vuosikate, euroa/asukas, 2017', 'Lainakanta, euroa/asukas, 2017',
       'Konsernin lainakanta, euroa/asukas, 2017',
       'Opetus- ja kulttuuritoiminta yhteensä, nettokäyttökustannukset, euroa/asukas, 2017',
       'Sosiaali- ja terveystoiminta yhteensä, nettokäyttökustannukset, euroa/asukas, 2017']
```

```python
#Viisi kuntaa suodatettuna, joiden työllisyysaste prosentteina on pienin.
df[["Kunta", "Työllisyysaste, %, 2016"]].sort_values(by="Työllisyysaste, %, 2016").head(5)
```

```python
#Viisi kuntaa suodatettuna, joiden työllisyysaste prosentteina on suurin.
df[["Kunta", "Työllisyysaste, %, 2016"]].sort_values(by="Työllisyysaste, %, 2016", ascending=False).head(5)
```

```python
#Kahdeksan kuntaa, joissa eläkeläisten prosenttiosuus on pienin, järjestyksessä pienimmästä alkaen.
df[["Kunta", "Eläkeläisten %-osuus väestöstä 2016"]].sort_values(by="Eläkeläisten %-osuus väestöstä 2016").head(8)                                                                                                   
```

```python
#Kymmenen väkiluvultaan suurinta kuntaa järjestyksessä suurimmasta pienimpään.
df[["Kunta", "Väkiluku 2017"]].sort_values(by="Väkiluku 2017", ascending=False).head(10)
```

```python
#Kymmenen kunnan ikärakenne, joissa yli 64 vuotiaiden prosenttiosuus väestöstä on suurin.
df[["Kunta", 'Alle 15v %-osuus väestöstä 2017', '15-64v %-osuus väestöstä 2017', 'Yli 64v %-osuus väestöstä 2017']].sort_values(by="Yli 64v %-osuus väestöstä 2017", ascending=False).head(10)
```

```python
#Kymmenen pienimmän työttömyysprosenttiosuuden omaavaa kuntaa.
df[["Kunta", "Alueella asuvan työllisen työvoiman määrä, 2016", "Alueella olevien työpaikkojen lukumäärä 2016", "Työttömien %-osuus työvoimasta 2016"]].sort_values(by="Työttömien %-osuus työvoimasta 2016").head(10)
```

```python
#Kahdeksan kuntaa järjestyksessä sen mukaisesti, joissa yli 15 vuotiaiden osuus väestöstä on suurin.
df["Yli 15v osuus väestöstä"] = df[['15-64v %-osuus väestöstä 2017' ,'Yli 64v %-osuus väestöstä 2017']].sum(axis=1)
df.head(5)
df[["Kunta", "Yli 15v osuus väestöstä"]].sort_values(by='Yli 15v osuus väestöstä', ascending=False).head(8)
```

```python
#Ne kunnat suodatettuna, joissa ruotsinkielisten prosenttiosuus väestöstä on yli 90%.
df[["Kunta","Ruotsinkielisten %-osuus väestöstä 2017"]][df["Ruotsinkielisten %-osuus väestöstä 2017"]>=90].sort_values(by="Ruotsinkielisten %-osuus väestöstä 2017")
```

```python
#Ne kunnat suodatettuna, joiden väkiluku on yli 100 000 ja työllisyysaste prosentteina alle 70.
df[["Kunta","Työllisyysaste, %, 2016", "Väkiluku 2017"]][(df["Työllisyysaste, %, 2016"]<70)&(df["Väkiluku 2017"]>100000)].sort_values(by="Työllisyysaste, %, 2016", ascending=True)
```
