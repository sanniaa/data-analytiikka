```python
#Tarvittavien kirjastojen tuominen.
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Datan avaaminen nettisivulta.
df = pd.read_csv('http://taanila.fi/BlackFriday.csv')
```

```python
#Muuttujien nimien muuttaminen selkeämmiksi.
df.columns = ['Tilausnro', 'Tuotenro', 'Sukupuoli', 'Ikä', 'Ammatti', 'Kaupunki',
       'Asumisaika', 'Siviilisääty', 'Tuote1',
       'Tuote2', 'Tuote3', 'Ostokset dollareina']
df.head(5)
```

```python
#Listojen tekeminen.
Siviilisääty = ["Naimaton", "Naimisissa"]
Sukupuoli = ["Mies", "Nainen"]
```

```python
#Sukupuolijakauma Blackfridayn ostosten tekijöistä.
sns.countplot(df["Sukupuoli"], palette= sns.color_palette("husl", 4)).set_title("Blackfridayn ostosten tekijöiden sukupuoli")
```

```python
#Kuvaus siitä, minkä ikäisiä Blackfridayna ostosten tekijät ovat olleet.
sns.countplot(df["Ikä"], palette= sns.color_palette("husl", 7)).set_title("Blackfriday ikäjakauma")
```

```python
#Yhdistettynä edelliset tekijät, eli iän ja sukupuolen perusteella kuvattu Blackfridayn ostajakuntaa.
sns.countplot(df["Ikä"],hue=df["Sukupuoli"], palette= sns.color_palette("husl", 2)).set_title("Blackfriday sukupuoli- ja ikäjakauma")
```

```python
#BlackFriday-ostosten tekijöiden lukumäärät kaupungin ja sukupuolen mukaisesti.
sns.countplot(x='Kaupunki', hue='Sukupuoli', data=df, palette= sns.color_palette("Set2", 2))
```

```python
#BlackFriday-ostosten tekijöiden jakauma iän ja kaupungin perusteella.
sns.countplot(x='Ikä', hue='Kaupunki', data=df, palette= sns.color_palette("husl", 5))
```

```python
#Ostosten suuruus esitetty ikäluokittain laatikko-janakuvioilla.
plt.style.use('grayscale')
ax = df.boxplot("Ostokset dollareina", by = "Ikä")
plt.title('Ostosten suuruudet ikäluokan mukaisesti')
plt.suptitle('')
plt.xlabel("Ikä")
plt.ylabel("Ostokset dollareina")
```

```python
#Ostosten tekijöiden jakauma kaupungeittain.
colors = ["lightyellow", "lightblue", "lightgreen"]
fig1, ax1 = plt.subplots(figsize=(12,7))
ax1.pie(df["Kaupunki"].value_counts(), labels=df["Kaupunki"].unique(), autopct='%1.1f%%', colors = colors) 
plt.tight_layout()
```

```python
#Ostosten tekijöiden jakauma ammattiluokittain.
fig1, ax1 = plt.subplots(figsize=(12,7))
colors = ["C2", "C3", "C4", "C5", "C6"]
ax1.pie(df["Ammatti"].value_counts(), labels=df["Ammatti"].unique(), autopct='%1.0f%%', colors = colors) 
plt.tight_layout()
```
